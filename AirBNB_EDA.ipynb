```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Load the Dataset

```python
file_path = "/content/datasets.csv"
data = pd.read_csv(file_path, encoding_errors='ignore')
```

Data Cleaning

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.shape
```

```python
#Duplicate Rows
data[data.duplicated()]
```

```python
#deleting duplicate rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data.dtypes
```

```python
# changing data types
data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

**Data Analysis**

```python
# PRICE
df=data[data['price']<1500]
sns.boxplot(data=df, x='price')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show
```

```python
df.columns
```

```python
#availability_365 Distribution
plt.figure(figsize=(5,5))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show
```

```python
#Price per bed
df['price per bed']=df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
df.columns
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price',hue='room_type')
```

```python
# Relationshp btwn no of reviwes & price
sns.scatterplot(data=df, x='number_of_reviews',y='price', hue='neighbourhood_group')
plt.title('Relationship between No.of Reviews & Price')
plt.figure(figsize=(8,5))

```

```python
df.dtypes
```

```python
sns.pairplot(data=df,vars=('price','minimum_nights','number_of_reviews','availability_365'), hue='room_type')
```

```python
#Geographical Understanding of AirBnb Listing
plt.figure(figsize=(10,7))
plt.title("Geographical Understanding of AirBnb Listing")
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.show()
```

```python
#Correction of one variable  with other variable
corr=df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
# Heatmap
sns.heatmap(data=corr, annot=True)
plt.figure(figsize=(8,5))
```
