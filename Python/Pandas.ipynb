```python
# read csv file
csv=pd.read_csv('/Chess trend data.csv')
csv.head()# 1st 5 rows/records
csv.tail()#last 5 records
csv.tail(20)#last 20 records
csv.Age.drop_duplicates()#drop duplicates & give unique values in a Age column
csv.describe() #describe the table
csv.Age.describe()#describe age column
csv.ELO.value_counts() # count the values
csv.Age.value_counts() # count the values of Age
csv.Age.value_counts(normalize=True) # Relative frequency of each unique value in the age column...
#the proportion of each value relative to the total number of values)
type(csv.Age.value_counts())
csv.Age.unique() # print unique values in age column as anarray
print(list(csv.Age.unique()))# print unique values in age column as a list
csv.ELO.nunique() # number of unique values in ELO column
csv.Age.nunique() # number of unique values in age column

```

```python
#Crosstab
pd.crosstab(csv.Age,csv.ELO)
```

```python
#Std
csv.Age.mean()#mean
csv.Age.median()#median
csv.Age.mode()#mode
csv.Age.plot(kind='hist')
csv.ELO.plot(kind='hist')
```
