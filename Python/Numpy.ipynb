```python
#1D Array
n1=np.array([1,2,3])
print('n1=',n1)

#2D Array
n2=np.array([[1,2,3],[4,5,6]])
print('n2=',n2)

#Zero filled array
n3=np.zeros([])
print('n3=',n3)

#zero filled array for 3 rows & 5 columns
n4=np.zeros([3,5],dtype=int)
print('n4=',n4)

#ones filled array
n5=np.ones([2,3],dtype=int)
print('n5=',n5)

#Empty array random / junk values
n6=np.empty([2,4])
print('n6=',n6)

#create an with a range of values
n7=np.arange(5)
print('n7=',n7)

#linespace
n8=np.linspace(1,20,num=5, dtype=int)
print('n8=',n8)

#full
n9=np.full([2,5],67)
print('n9=',n9)

#fill
n5.fill(9)
print('n5=',n5)
```

```python
#Accesssing Array
print(n1)
print(n1[0]) # print 1st element
print(n1[-1]) # print last element
print(n2)
print(n2[1,0]) #n2[0]-1st list n2[1]-2nd list
print(n2[1,:]) # print evrything in the 1st list
print(n2[:,1]) # print evrything in the 2nd column
n4=np.ones([2,3,4],dtype=int)
print(n4)
n4[1,1,2]=10
print(n4)
```

```python
#Size & Shape
print(n4.ndim) # dimension
print(n4.size) # size or no of elements
print(n4.shape) # arryas,row,columns we have
print(n4.dtype) # data type
print(n4.itemsize) # byte size of one element
print(n4.nbytes) # total byte size
```

```python
#sorting
n10=np.array([7,2,9,10,1,6,3,4,15])
np.sort(n10)

n11=np.array([[[7,2,3],[3,15,0]],[[2,99,25],[100,0,19]]])
np.sort(n11,axis=0)

```

```python
# Reshaping
print(n10)
n12=n10.reshape(3,3)
print(n12)
```

```python
#Joining Array
n13=np.arange(4)
n14=np.arange(5,9)
n15=np.concatenate([n13,n14])
print(n13)
print(n14)
print(n15)
```

```python
#Converting 1D to 2D
n16=np.arange(8)
print(n16)
n17=n16[np.newaxis,:]
print(n17)
print(n17.shape)
n18=n16[:,np.newaxis]
print(n18)
print(n18.shape)
n19=np.expand_dims(n16,axis=0)
print(n19)
n20=np.expand_dims(n16,axis=1)
print(n20)
```

```python
#Indexing & Slicing
n21=np.arange(10)
print(n21)
print(n21[(n21>3) & (n21<8)])
grater_than_five=n21>5
print(grater_than_five)
print(n21[grater_than_five])
print(n21[3:7])
n21[6]=100
print(n21[3:7])

```

```python
#stack & split
n22=np.array([[1,2],[3,4]])
n23=np.array([[5,6],[7,8]])
np.hstack((n22,n23))
np.vstack((n22,n23))
n24=np.arange(20).reshape(4,5)
print(n24)
np.hsplit(n24,5)
np.vsplit(n24,4)
```

```python
n24=np.arange(20).reshape(4,5)
print(n24)
np.hsplit(n24,5)
np.vsplit(n24,4)
```

```python
#copy
n25=n24
print(n24)
print(n25)
n24[0]=100
print(n24)
print(n25)
```

```python
#shallow copy
n25=n24[0:5]
n24[0]=100
print(n24)
print(n25)
```

```python
#deep copy
import copy
n26=copy.deepcopy(n24)


print(n26)
print(n24)

```

```python
#Math Functions
print(n26.sum())
print(n26.max())
print(n26.min())
print(n26*10)
np.cos(n26)
```

```python
#Random Functions
np.random.rand(3,3)
n27=np.random.randint(100,size=(4,2))
n27
```

```python
print(np.sqrt(n27))
print(np.exp(n27))
```

```python
#multiplication
m1=np.array([[1,2],[3,4]])
m2=np.array([[5,2],[3,6]])
print(m1)
print(m2)

m3=np.dot(m1,m2)
print(m3)


#inverse matrix
n4=np.linalg.inv(m1)
print(n4)
```
